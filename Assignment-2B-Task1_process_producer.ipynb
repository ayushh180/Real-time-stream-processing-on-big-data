```python
# import statements
from time import sleep
import time
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt
import csv


def read_csv(fileName):
    '''Read the CSV file clickstream.csv'''
    '''you can use csv.DictReader'''
    data = []
    file_data = csv.DictReader(open("Streaming_Linux_process.csv"))
    for row in file_data:
        data.append(row)
    
    return data

def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, data)
        print('Message published successfully. Data: ' + str(data))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer
    
    
# functions to define data schema    
def dataType(i):
    i['sequence'] = int(i['sequence'])
    i['machine'] = int(i['machine'])
    i['PID'] = int(i['PID'])
    i['TRUN'] = int(i['TRUN'])
    i['TSLPI'] = int(i['TSLPI'])
    i['TSLPU'] = int(i['TSLPU'])
    i['POLI'] = str(i['POLI'])
    i['NICE'] = int(i['NICE'])
    i['PRI'] = int(i['PRI'])
    i['RTPR'] = int(i['RTPR'])
    i['CPUNR'] = int(i['CPUNR'])
    i['Status'] = str(i['Status'])
    i['EXC'] = int(i['EXC'])
    i['State'] = str(i['State'])
    i['CPU'] = float(i['CPU'])
    i['CMD'] = str(i['CMD'])
    return i
    
if __name__ == '__main__':
   
    topic = 'NewLinuxProcessStream2B'
    cRows = read_csv('Streaming_Linux_process.csv')
    
    print('Publishing records..')
    producer = connect_kafka_producer()
    
    utc_date = dt.datetime.utcnow()
    unix_timestamp = (int(time.mktime(utc_date.timetuple())))
    # seperating the data based on machine
    machine4=[dataType(x) for x in cRows if x['machine'] == '4']
    machine5=[dataType(x) for x in cRows if x['machine'] == '5']
    machine6=[dataType(x) for x in cRows if x['machine'] == '6']
    machine7=[dataType(x) for x in cRows if x['machine'] == '7']
    machine8=[dataType(x) for x in cRows if x['machine'] == '8']
    
    
    initial_no4 = 0
    initial_no5 = 0
    initial_no6 = 0
    initial_no7 = 0
    initial_no8 = 0
    
    # looping to stream the data for live streaming
    while True:       
    # introducting random no for range for which data will be streamed and adding time as UTC and adding timestamp
        data=[]
        random_no4 = initial_no4 + random.randint(10,50)
        random_no5 = initial_no5 + random.randint(10,50)
        random_no6 = initial_no6 + random.randint(10,50)
        random_no7 = initial_no7 + random.randint(10,50)
        random_no8 = initial_no8 + random.randint(10,50)
        utc_date = dt.datetime.utcnow()
        unix_timestamp = (int(time.mktime(utc_date.timetuple())))
# condition for machine 4 if random no range extends more that the range of data and adding timestamp
        if (random_no4)>len(machine4):
            data.extend(machine4[initial_no4:len(machine4)])
            restart4=(random_no4)-len(machine4)
            data.extend(machine4[0:restart4])
            initial_no4=restart4
            for i in data:
                i.update( { "ts": unix_timestamp } )
        else:
            data.extend(machine4[initial_no4:random_no4])
            initial_no4 = random_no4
            for i in data:
                i.update( { "ts": unix_timestamp } )
            
# condition for machine 5 if random no range extends more that the range of data and adding timestamp
        if (random_no5)>len(machine5):
            data.extend(machine5[initial_no5:len(machine5)])
            restart5=(random_no5)-len(machine5)
            data.extend(machine5[0:restart5])
            initial_no5=restart5
            for i in data:
                i.update( { "ts": unix_timestamp } )
        else:
            data.extend(machine5[initial_no5:random_no5])
            initial_no5 = random_no5 
            for i in data:
                i.update( { "ts": unix_timestamp } )

        # condition for machine 6 if random no range extends more that the range of data and adding timestamp
        if (random_no6)>len(machine6):
            data.extend(machine6[initial_no6:len(machine6)])
            restart6=(random_no6)-len(machine6)
            data.extend(machine6[0:restart6])
            initial_no6=restart6
            for i in data:
                i.update( { "ts": unix_timestamp } )
        else:
            data.extend(machine6[initial_no6:random_no6])
            initial_no6 = random_no6
         # condition for machine 7 if random no range extends more that the range of data and adding timestamp
        
        
        if (random_no7)>len(machine7):
            data.extend(machine7[initial_no7:len(machine7)])
            restart7=(random_no7)-len(machine7)
            data.extend(machine7[0:restart7])
            initial_no7=restart6
            for i in data:
                i.update( { "ts": unix_timestamp } )      
        
        else:
            data.extend(machine7[initial_no7:random_no7])
            initial_no7 = random_no7 
            for i in data:
                i.update( { "ts": unix_timestamp } )
        
        
        # condition for machine 8 if random no range extends more that the range of data and adding timestamp
        
        
        if (random_no8)>len(machine8):
            data.extend(machine8[initial_no8:len(machine8)])
            restart8=(random_no8)-len(machine8)
            data.extend(machine8[0:restart8])
            initial_no8=restart8
            for i in data:
                i.update( { "ts": unix_timestamp } )
                
        else:
            data.extend(machine8[initial_no8:random_no8])
            initial_no8 = random_no8 
            for i in data:
                i.update( { "ts": unix_timestamp } )
                
                
        #Publishing the data
        publish_message(producer, topic, data)
        sleep(5)  
```
