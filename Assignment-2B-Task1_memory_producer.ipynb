```python
# import statements
from time import sleep
import time
from json import dumps
from kafka import KafkaProducer
import random
import datetime as dt
import csv


def read_csv(fileName):
    '''Read the CSV file clickstream.csv'''
    '''you can use csv.DictReader'''
    data = []
    file_data = csv.DictReader(open("Streaming_Linux_memory.csv"))
    for row in file_data:
        data.append(row)
    
    return data

def publish_message(producer_instance, topic_name, data):
    try:
        producer_instance.send(topic_name, data)
        print('Message published successfully. Data: ' + str(data))
    except Exception as ex:
        print('Exception in publishing message.')
        print(str(ex))
        
def connect_kafka_producer():
    _producer = None
    try:
        _producer = KafkaProducer(bootstrap_servers=['localhost:9092'],
                                  value_serializer=lambda x: dumps(x).encode('ascii'),
                                  api_version=(0, 10))
    except Exception as ex:
        print('Exception while connecting Kafka.')
        print(str(ex))
    finally:
        return _producer

    # Defining data type
    
def dataType(data):
    data['sequence']=int(data.get('sequence'))
    data['machine']=int(data.get('machine'))
    data['PID']=int(data.get('PID'))
    data['MINFLT']=str(data.get('MINFLT'))
    data['MAJFLT']=str(data.get('MAJFLT'))
    data['VSTEXT']=str(data.get('VSTEXT'))
    data['VSIZE']=float(data.get('VSIZE'))
    data['RSIZE']=str(data.get('RSIZE'))
    data['VGROW']=str(data.get('VGROW'))
    data['RGROW']=str(data.get('RGROW'))
    data['MEM']=float(data.get('MEM'))
    data['CMD']=str(data.get('CMD'))
    return data


if __name__ == '__main__':
   
    topic = 'NewLinuxMemoryStream2B'
    cRows = read_csv('Streaming_Linux_memory.csv')
    
    print('Publishing records..')
    producer = connect_kafka_producer()
    
    # seperating the data based on machine
    
    
    machine4=[dataType(x) for x in cRows if x['machine'] == '4']
    machine5=[dataType(x) for x in cRows if x['machine'] == '5']
    machine6=[dataType(x) for x in cRows if x['machine'] == '6']
    machine7=[dataType(x) for x in cRows if x['machine'] == '7']
    machine8=[dataType(x) for x in cRows if x['machine'] == '8']

    
    # new list for each X and Y and a final data
    data=[]
    dataX=[]
    dataY=[]
# setting intial number for each machine and generating random number
    xinitial_no4=0
    xinitial_no5=0
    xinitial_no6=0
    xinitial_no7=0
    xinitial_no8=0

    xrandom_no4 = xinitial_no4 + random.randint(20,80)
    xrandom_no5 = xinitial_no5 + random.randint(20,80)
    xrandom_no6 = xinitial_no6 + random.randint(20,80)
    xrandom_no7 = xinitial_no7 + random.randint(20,80)
    xrandom_no8 = xinitial_no8 + random.randint(20,80)

    # Introducting time stamp
    utc_date = dt.datetime.utcnow()
    unix_timestamp = (int(time.mktime(utc_date.timetuple())))
    
    # Appending streaming data for X for first time
    dataX=machine4[xinitial_no4:xrandom_no4]
    dataX.extend(machine5[xinitial_no5:xrandom_no5])
    dataX.extend(machine6[xinitial_no6:xrandom_no6])
    dataX.extend(machine7[xinitial_no7:xrandom_no7])
    dataX.extend(machine8[xinitial_no8:xrandom_no8])
    for i in dataX:
        i.update( { "ts": unix_timestamp } )
    
    # publishing data and setting new index
    data=dataX
    publish_message(producer, topic, data)
    xinital_no4=xrandom_no4
    xinital_no5=xrandom_no5
    xinital_no8=xrandom_no6
    xinital_no7=xrandom_no7
    xinital_no8=xrandom_no8

    
    # introducting inital value for streaming data and generating random no for range
    yinitial_no4 = 0
    yinitial_no5 = 0
    yinitial_no6 = 0
    yinitial_no7 = 0
    yinitial_no8 = 0

# adding for range  and specifying the range for live streaming 
    yrandom_no4 = yinitial_no4 + random.randint(0,5)
    yrandom_no5 = yinitial_no5 + random.randint(0,5)
    yrandom_no6 = yinitial_no6 + random.randint(0,5)
    yrandom_no7 = yinitial_no7 + random.randint(0,5)
    yrandom_no8 = yinitial_no8 + random.randint(0,5)

    dataY=(machine4[yinitial_no4:yrandom_no4])
    dataY.extend(machine5[yinitial_no5:yrandom_no5])
    dataY.extend(machine6[yinitial_no6:yrandom_no6])
    dataY.extend(machine7[yinitial_no7:yrandom_no7])
    dataY.extend(machine8[yinitial_no8:yrandom_no8])
    for i in dataY:
        i.update( { "ts": unix_timestamp } )


# setting new initial value for y
    yinitial_no4 = yrandom_no4
    yinitial_no5 = yrandom_no5
    yinitial_no6 = yrandom_no6
    yinitial_no7 = yrandom_no7
    yinitial_no8 = yrandom_no8
    
    sleep(10)

# Looping to generate the in the following format specified in ques like 
# X1, X2Y1,X3Y2,X4Y3 and so on
    while True:

            xrandom_no4 = xinitial_no4 + random.randint(10,50)
            xrandom_no5 = xinitial_no5 + random.randint(10,50)
            xrandom_no6 = xinitial_no6 + random.randint(10,50)
            xrandom_no7 = xinitial_no7 + random.randint(10,50)
            xrandom_no8 = xinitial_no8 + random.randint(10,50)
            utc_date1 = dt.datetime.utcnow()
            unix_timestamp1= (int(time.mktime(utc_date1.timetuple())))


            if (xrandom_no4)>len(machine4):
                dataX=(machine4[xinitial_no4:len(machine4)])
                restart4=(xrandom_no4)-len(machine4)
                dataX.extend(machine4[0:restart4])
                xinitial_no4=restart4
            else:
                dataX=(machine4[xinitial_no4:xrandom_no4])
                xinitial_no4 = xrandom_no4


            if (xrandom_no5)>len(machine5):
                dataX.extend(machine5[xinitial_no5:len(machine5)])
                restart5=(xrandom_no5)-len(machine5)
                dataX.extend(machine5[0:restart5])
                xinitial_no5=restart5
            else:
                dataX.extend(machine5[xinitial_no5:xrandom_no5])
                xinitial_no5 = xrandom_no5 

            if (xrandom_no6)>len(machine6):
                dataX.extend(machine6[xinitial_no6:len(machine6)])
                restart6=(xrandom_no6)-len(machine6)
                dataX.extend(machine6[0:restart6])
                xinitial_no6=restart6
            else:
                dataX.extend(machine6[xinitial_no6:xrandom_no6])
                xinitial_no6 = xrandom_no6



            if (xrandom_no7)>len(machine7):
                dataX.extend(machine7[xinitial_no7:len(machine7)])
                restart7=(xrandom_no7)-len(machine7)
                dataX.extend(machine7[0:restart7])
                xinitial_no7=restart6
     

            else:
                dataX.extend(machine7[xinitial_no7:xrandom_no7])
                xinitial_no7 = xrandom_no7 


            if (xrandom_no8)>len(machine8):
                dataX.extend(machine8[xinitial_no8:len(machine8)])
                restart8=(xrandom_no8)-len(machine8)
                dataX.extend(machine8[0:restart8])
                xinitial_no8=restart8


            else:
                dataX.extend(machine8[xinitial_no8:xrandom_no8])
                xinitial_no8 = xrandom_no8 
            for i in dataX:
                i.update( { "ts": unix_timestamp1 } )


            data=dataX+dataY
            publish_message(producer, topic, data)
            
            
            
            xinital_no4=xrandom_no4
            xinital_no5=xrandom_no5
            xinital_no8=xrandom_no6
            xinital_no7=xrandom_no7
            xinital_no8=xrandom_no8
            
            
            yrandom_no4 = yinitial_no4 + random.randint(0,5)
            yrandom_no5 = yinitial_no5 + random.randint(0,5)
            yrandom_no6 = yinitial_no6 + random.randint(0,5)
            yrandom_no7 = yinitial_no7 + random.randint(0,5)
            yrandom_no8 = yinitial_no8 + random.randint(0,5)
            

            if (yrandom_no4)>len(machine4):
                dataY=(machine4[yinitial_no4:len(machine4)])
                restart4=(yrandom_no4)-len(machine4)
                dataY.extend(machine4[0:restart4])
                yinitial_no4=restart4

            else:
                dataY=(machine4[yinitial_no4:yrandom_no4])
                yinitial_no4 = yrandom_no4


            if (yrandom_no5)>len(machine5):
                dataY.extend(machine5[yinitial_no5:len(machine5)])
                restart5=(yrandom_no5)-len(machine5)
                dataY.extend(machine5[0:restart5])
                yinitial_no5=restart5

            else:
                dataY.extend(machine5[yinitial_no5:yrandom_no5])
                yinitial_no5 = yrandom_no5 


            if (yrandom_no6)>len(machine6):
                dataY.extend(machine6[yinitial_no6:len(machine6)])
                restart6=(yrandom_no6)-len(machine6)
                dataY.extend(machine6[0:restart6])
                yinitial_no6=restart6

            else:
                dataY.extend(machine6[yinitial_no6:yrandom_no6])
                yinitial_no6 = yrandom_no6



            if (yrandom_no7)>len(machine7):
                dataY.extend(machine7[yinitial_no7:len(machine7)])
                restart7=(yrandom_no7)-len(machine7)
                dataY.extend(machine7[0:restart7])
                yinitial_no7=restart6    

            else:
                dataY.extend(machine7[xinitial_no7:xrandom_no7])
                yinitial_no7 = yrandom_no7 


            if (yrandom_no8)>len(machine8):
                dataY.extend(machine8[yinitial_no8:len(machine8)])
                restart8=(yrandom_no8)-len(machine8)
                dataY.extend(machine8[0:restart8])
                yinitial_no8=restart8

            else:
                dataY.extend(machine8[yinitial_no8:yrandom_no8])
                yinitial_no8 = yrandom_no8 
            
            
            for i in dataY:
                i.update( { "ts": unix_timestamp1 } )
                
                
                
                
            yinitial_no4 = yrandom_no4
            yinitial_no5 = yrandom_no5
            yinitial_no6 = yrandom_no6
            yinitial_no7 = yrandom_no7
            yinitial_no8 = yrandom_no8
# wait for 10 sec for next batch
            sleep(10)

```
